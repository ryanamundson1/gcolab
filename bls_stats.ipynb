```
#This is using the BLS pip package: https://pypi.org/project/bls/

import bls
import seaborn as sns
import matplotlib.pyplot as plt

# Get Obama years and update to today's date for comparison
obStats = bls.api.get_series('LNS12000000', 2012, 2016)
obStats.index = obStats.index.to_timestamp() + pd.DateOffset(years=4)


trStats = bls.api.get_series('LNS12000000', 2017, 2019)
```

```
# Cross sectional of just manufacturing Obama years and update to today's date for comparison
obStatsMan = bls.api.get_series('CES3000000001', 2012, 2016)
obStatsMan.index = obStatsMan.index.to_timestamp() + pd.DateOffset(years=4)


trStatsMan = bls.api.get_series('CES3000000001', 2017, 2019)
```

```
import pandas as pd
import numpy as np


fig, ax1 = plt.subplots()

plt.figure(figsize=(30,20))

ax2 = ax1.twinx()

color = 'tab:blue'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('Obama Jobs', color=color)

opPlot = obStats.plot(kind='line', ax=ax1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

minR = min(obStats)
maxR = max(obStats)

rangeR = maxR - minR

tRange = min(obStats)


color = 'tab:red'
ax2.set_ylabel('Trump Jobs', color=color)
trPlot = trStats.plot(kind='line', ax=ax2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
trPlot.set_yticks(np.arange(min(trStats) - 128, min(trStats)+rangeR, 2000))
print (trPlot.get_yscale())
plt.show()
```
